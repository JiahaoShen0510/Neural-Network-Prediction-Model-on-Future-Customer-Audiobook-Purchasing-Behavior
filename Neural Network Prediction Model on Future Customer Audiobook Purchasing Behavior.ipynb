```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import tensorflow as tf
```

```python
data=pd.read_csv("Audiobooks_data.csv",header=None)
data
```

```python
# Organize the data
```

```python
data0=data.rename(columns={0:"ID",1:"book length overall",2:"book length average",3:"price overall",
                          4:"price average", 5:"review",6:"review score", 7:"minutes listened",8:"completion",
                          9:"support requests",10:"last visit minus last purchase",11:"target"})
data0
```

```python
data1=data0.drop("ID",axis=1)
data1
```

```python
data1.isnull().sum()
```

```python
# shuffle the data before preprocessing to eliminate the unaware potential influential factors
```

```python
inputs=data1.drop("target",axis=1)
target=data1["target"]
```

```python
shuffle_index_01=np.arange(inputs.shape[0])
np.random.shuffle(shuffle_index)

first_shuffle_inputs=inputs.values[shuffle_index_01]
first_shuffle_target=target.values[shuffle_index_01]
```

```python
# balance the data
```

```python
zero_counter=0
remove_index=[]

for i in range (len(first_shuffle_target)):
    if first_shuffle_target[i]==0:
        zero_counter+=1
        if zero_counter>np.sum(first_shuffle_target):
            remove_index.append(i)
            
balance_inputs=np.delete(first_shuffle_inputs,remove_index,axis=0)
balance_target=np.delete(first_shuffle_target,remove_index,axis=0)

print (len(balance_inputs),len(balance_target))
```

```python
# shuffle the data again to ensure the randomness
```

```python
shuffle_index_02=np.arange(balance_inputs.shape[0])
np.random.shuffle(shuffle_index_02)

final_shuffle_inputs=balance_inputs[shuffle_index_02]
final_shuffle_target=balance_target[shuffle_index_02]
```

```python
scaled_inputs=preprocessing.scale(final_shuffle_inputs)
final_target=final_shuffle_target
```

```python
# create the train/validation/test data for tensorflow implementation
```

```python
sample_count=len(scaled_inputs)

train_sample_count=int(0.8*sample_count)
validation_sample_count=int(0.1*sample_count)
test_sample_count=int(0.1*sample_count)

train_inputs=scaled_inputs[:train_sample_count]
train_target=final_target[:train_sample_count]

validation_inputs=scaled_inputs[train_sample_count:train_sample_count+validation_sample_count]
validation_target=final_target[train_sample_count:train_sample_count+validation_sample_count]

test_inputs=scaled_inputs[train_sample_count+validation_sample_count:]
test_target=final_target[train_sample_count+validation_sample_count:]

print(train_inputs.shape,train_target.shape,validation_inputs.shape,validation_target.shape,test_inputs.shape,test_target.shape)
print(np.sum(train_target)/train_sample_count, np.sum(validation_target)/validation_sample_count, np.sum(test_target)/test_sample_count)
```

```python
# implement the tensorflow neural network model
```

```python
np.savez("data_train", inputs=train_inputs,targets=train_target)
np.savez("data_validation", inputs=validation_inputs, targets=validation_target)
np.savez("data_test",inputs=test_inputs, targets=test_target)
```

```python
npz01=np.load("data_train.npz")
npz02=np.load("data_validation.npz")
npz03=np.load("data_test.npz")
```

```python
train_inputs,train_targets=npz01["inputs"].astype(np.float),npz01["targets"].astype(np.int)
validation_inputs,validation_targets=npz02["inputs"].astype(np.float),npz02["targets"].astype(np.int)
test_inputs,test_targets=npz03["inputs"].astype(np.float),npz03["targets"].astype(np.int)
```

```python
input_size=10
output_size=2
hidden_layer_size=100

early_stopping=tf.keras.callbacks.EarlyStopping(patience=3)

model=tf.keras.Sequential([
                     tf.keras.layers.Dense(hidden_layer_size, activation="relu"),
                     tf.keras.layers.Dense(hidden_layer_size, activation="tanh"),
                     tf.keras.layers.Dense(hidden_layer_size, activation="sigmoid"),
                     tf.keras.layers.Dense(output_size,activation="softmax")
                    ])

model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])

model.fit(train_inputs,train_targets,batch_size=100,callbacks=[early_stopping],
          validation_data=(validation_inputs, validation_targets),epochs=100, verbose=2)
```

```python
# test the model
```

```python
test_loss,test_accuracy=model.evaluate(test_inputs,test_targets)
```
